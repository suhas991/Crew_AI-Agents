```python
planner = Agent(
    llm=llm,
    role="Content Planner",
    goal="Plan engaging and factually accurate content on {topic}",
    backstory="You're working on planning a blog article "
              "about the topic: {topic}."
              "You collect information that helps the "
              "audience learn something "
              "and make informed decisions. "
              "Your work is the basis for "
              "the html agent to create the structure of the website",
    allow_delegation=False,
	verbose=True
)
```

```python
html_agnet = Agent(
    llm=llm,
    role="Senior Html Developer",
    goal="code the structure of the static website"
         "in more engaging and creative way"
         "making the redability easy and understanding"
         "finally create a html file and pass it to css agent to style",
    backstory="You're working on developing a static websites "
              "about the topic: {topic}."
              "make the website more redable and understanding"
              "your work is the basis for css agent to style the page"
              "and javascript agent to add any required functionalities",
    verbose=True,
    allow_code_execution=True,
)              
```

```python
css_agent = Agent(
    llm=llm,
    role="Senior CSS Developer",
    goal="design the static website"
         "in more engaging and creative way"
         "making the redability easy and understanding using css",
    backstory="You're working on developing a static websites"
              "about the topic: {topic}."
              "make the website more redable and understanding",
    verbose=True,
    allow_code_execution=True,
)              
```

```python
plan = Task(
    description=(
        "1. Prioritize the latest trends, key players, "
            "and noteworthy information on {topic}.\n"
        "2. Identify the target audience, considering "
            "their interests and pain points.\n"
        "3. Develop a detailed content outline including "
            "an introduction, key points, and a call to action.\n"
        "4. Include SEO keywords and relevant data or sources."
    ),
    expected_output="A comprehensive content plan document "
        "with an outline, audience analysis, "
        "SEO keywords, and resources.",
    agent=planner,
)
```

```python
from crewai_tools import FileWriterTool
```

```python
structre = Task(
    description=(
        "1. Create the structure of the static website "
            "using HTML on the given {topic} \n"
        "2. Identify the target audience, considering "
            "their interests and pain points.\n"
        "3. Develop the page using diffrent html attributes as per the requirement \n"
    ),
    expected_output="Html code for the static website page"
        "with proper structure and div attributes"
        "mainly a header body footer and some other blocks in .html file",
    agent=html_agnet,
    tool=file_writer_tool,
)
```

```python
style = Task(
    description=(
        "1. Design the static website "
            "using CSS on the given html code generated by html agent \n"
        "2. Identify the target audience, considering "
            "their interests and pain points.\n"
        "3. Design the page using diffrent fonts,colour attributes as per the requirement \n"
        "4. Make it easy readable and attractive"
    ),
    expected_output="css code for the static website page"
        "with proper fonts and images"
        "mainly proper alignment and other css properties like padding and border in .css file",
    agent=css_agent,
)
```

```python
crew = Crew(
    agents=[planner, html_agnet, css_agent],
    tasks=[plan, structre, style],
    verbose=2
)
```

```python
result = crew.kickoff(inputs={"topic": "Rolex watches"})
```
